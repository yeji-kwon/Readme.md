우리 삶 속에서 인공지능은 떼려야 뗄 수 없는 기능이 되었다. 그중 가장 대표적인 언어, 음성, 이미지, 자율 주행이며 이 네 가지 분야에서 사용되는 제품과 적용되는 기술들을 분석해보고 알아보려고 한다.

**1. 언어**   
인공지능 번역은 AI 도입으로 인해 가장 큰 변화가 기대되는 분야 중 하나이다. ‘과연 미래에 인간이 외국어를 배울 필요가 있을까?’ 이 질문에 대해 지금은 확실하게 정답을 말할 수는 없지만 몇몇 전문가들은 미래에 언어를 따로 배우지 않더라도 세계 여러 사람들과 의사소통을 하는데 무리가 없을 것이라고 조심스레 추측한다. 수많은 번역 프로그램들이 많이 나와있지만 번역의 선두자라고 할 수 있는 구글 번역기에 대해 알아보고자 한다.

*-신경망 번역 기술*   
구글 번역기는 ‘신경망 번역’ 기술을 사용하고 있다. ‘신경망 번역’이란 인공지능을 기반으로 한 기계학습 알고리즘을 적용한 기술이다. 기존의 인공지능은 단어로 하나하나 번역했기 때문에 문장이 어색했지만 이젠 문장을 통째로 번역해서 맥락으로 그 뜻을 이해할 수 있다. 과거 구문 기반 번역 일 때는 기술의 한계가 뚜렷했으나 ‘신경망 번역’기술을 도입한 이후로는 번역이 엄청나게 향상했다. 또한 이용자가 늘수록 데이터가 축적되어 점점 번역 실력도 향상된다. 최근 구글 번역기는 무려 하루에 10억 개 이상의 문장과 1400억 개 이상의 단어를 번역한다. 또한 103개 언어를 지원하며 실사용자 수는 5억 명에 달하며 어마어마한 데이터를 자랑한다. 

*-웹 크롤링*   
구글 번역기가 사용하는 또 다른 기술 중 하나는 바로 ‘웹 크롤링’이다. 인공지능을 학습시키기 위해선 데이터를 많이 모으는 것이 관건이다. 보통 한 쌍의 언어, 예를 들어 한국어와 영어의 번역을 훈련하기 위해서는 약 1억 개 이상의 학습 사례가 필요하다. 이를 위해서 데이터를 수집하는 것을 ‘웹 크롤링’이라 한다. 데이터의 수집을 통해 데이터를 최신 상태로 유지하며 수많은 컴퓨터에 분산되어 있는 문서를 수집해 대상의 색인으로 포함하는 기술이다. 이를 통해 수십억 건의 데이터를 모아 인공지능에 학습시킨다. ‘웹 크롤링’은 번역 모델을 더욱 고도화하는데 꼭 필요한 기술이다.

*-알고리즘 작업과 다중언어 모델*   
방대한 데이터가 있다면 그 데이터를 잘 활용할 수 있는 기술의 정교함 또한 필요하다. 그래서 구글은 정교하고 세심한 알고리즘을 만드는데 집중하고 있다. 많은 데이터를 수집하더라도 그것이 좋은 번역인지 나쁜 번역인지 판단하는 알고리즘이 필요하고 그러한 알고리즘을 만드는 것은 단순한 작업이 아니다. 그래서 더욱 정교한 알고리즘을 위해 언어 전문가와 함께 힘쓰고 있다. 또한 여러 언어를 묶을 수 있는 ‘다중언어 모델’을 도입하였다. 다중언어 모델이란 ‘영어-한국어’의 번역을 통하여 ‘한국어-일본어’의 번역까지 실현하는 것이다. 이처럼 구글은 많은 기술을 통해 더 매끄럽고 편리한 번역을 위해 힘쓰고 있다.

**2. 음성**   
인공지능 음성 인식 기술은 우리가 일상에서 너무나 자연스럽게 사용하고 있다. 말 한마디로 오늘의 날씨를 알 수 있고 좋아하는 음악도 틀 수 있으며 운전 중에 전화와 메시지도 해결할 수 있다. 이처럼 음성 인식 기술은 이미 우리 삶에 깊이 스며 들었고 더욱 발전하고 있는 중이다. 인공지능 음성 서비스를 알아보던 중 수많은 음성 서비스 중에서도 내 눈을 확 사로잡은 기술이 있었다. 그것은 바로 셀바스 AI가 선보인 감정 표현 기능이었다. 물론 AI는 감정이 없지만 인간의 행복 슬픔과 같은 감정을 설정함으로 인공지능의 목소리를 다르게 하여 우리가 그 목소리를 들을 때 인공지능의 감정을 조절할 수 있다고 하는 기술이다. 실제로 셀바스에서 내놓은 음성합성 기술의 영상을 보았는데 굉장히 흥미로웠다. 단순히 행복, 슬픔, 다정의 감정을 선택하는 것에서 끝나는 것이 아니라 그 단계 또한 조절할 수 있어서 목소리의 떨림과 톤 신남의 정도까지 세심하게 달라졌다. 인공지능의 목소리가 인간과 굉장히 비슷해서 심지어 나와 감정을 공유하고 있다고까지 느껴졌다. 이러한 다양한 감정 표현 기능은 앞으로 뉴스, 교육 영상, 드라마 더빙 등 수많은 콘텐츠에 굉장히 유용하게 쓰일 것으로 기대가 된다. 


인공지능을 기반으로 한 음성 인식 제품에게 말을 하면 크게 세 가지의 과정이 이루어진다.    
*1) 음성 입력 및 인식*   
첫 번째 과정에서는 사람의 음성이 컴퓨터가 이해할 수 있도록 음성 입력을 거쳐 텍스트화한다. 이때 ‘음성합성(TTS=text to speech)’ 기술이 사용된다. 이 과정에선 주변의 소음을 다 제외하고 말소리만을 파악하여 각 각의 단어와 발음을 인식하여 입력하므로 생각보다 복잡한 과정이다.    
*2)자연어 처리 과정*    
자연어란 인간이 의사소통을 하기 위해 사용하는 언어이며 자연어 처리 과정은 컴퓨터가 자연어를 분석하여 처리하는 과정이다. 
이 과정에선 컴퓨터가 자연어에 대한 구문과 형태소 그 의미 등을 해석하고 분석하여 문장에 담긴 의미를 파악한다. 이 과정이 바로 AI가 일하는 주요 분야이다.    
*3) 인식 결과 출력*  
앞의 결과에서 인식된 결괏값에 따라 최적의 결과를 내놓고 ‘TTS(text to speech)’ 기술을 통해 마치 사람처럼 말소리로 응답을 한다. 날씨를 물어봤을 경우 인터넷에 검색하여 대답을 해주고 친구에게 전화를 하라고 했을 경우 전달한 내용에 따라 그 사람에게 연결해 준다. 

이렇게 간단히 세 가지의 과정으로 현재 음성 인식 기술에 대해 알아보았다. 음성 인식 기술은 단기간에 굉장히 많은 성과를 이루어 냈지만 한계도 보인다. 주변 소음으로 잘 인식하지 못하는 경우도 있고 사투리와 사람 고유의 억양 또한 잡아내지 못하는 경우도 많다. 음성 인식 기술이 더욱 정확성 있게 향상되기 위해선 앞으로 더 많은 데이터가 필요할 것 같다. 그렇게 된다면 더 다양한 분야에서 유용하게 활용될 수 있을 거라 기대도 해본다.   

**3. 이미지**  
최근 딥러닝을 통해 사람 못지않은 정확성을 가지고 다양한 사물을 인식할 수 있는 AI 이미지 분석이 등장하였다. 이미지 분석에 있어서 핵심 기술은 바로 그 이미지의 특징을 얼마나 잘 포착하고 추상화하여 인지하느냐에 있다. 인공지능 이미지 인식은 인공신경망으로 추상화와 특징 포착을 동시에 진행한다. 
요즘은 분석을 뛰어넘어 이미지를 생성하는 기술도 발전하고 있다. 오늘 함께 알아보고 싶은 서비스는 바로 ‘SC-FEGAN’이다. 전문 프로그램이 없더라도 쉽게 편집할 수 있는 기술인데 이 기술을 활용하면 간단한 낙서로도 쉽게 원하는 내용을 그려서 복원할 수 있으며 머리 모양이나 표정도 바꿀 수 있다.

*-GAN기능*   
이 서비스엔 딥러닝의 기법 중 하나인 GAN(Gernative Adversarial Network)기술이 사용된다. GAN 기술은 이미지를 변화하는데 탁월하며 해당 사진과 자연스럽게 어울리는 이미지로 만들 수 있다. 이러한 기술을 사용한다면 앞으로 디자인 분야에서 시간을 줄이면서 품질은 향상시킬 수 있을것으로 예상된다.

**4. 자율 주행**   
자동차가 스스로 운전하는 것은 십 년 전까지만 하더라도 상상에 불과했다. 그러나 자율 주행은 현재 이루어지고 있으며 머지않아 상용화 될 것이란 기대도 해본다.
현대기아차는 운전자의 성향에 맞도록 AI 기반 부분 자율 주행을 실현하는 SCC-ML기술을 세계 최초로 개발하고 특허를 냈다.    

*-SCC-ML기능*    
이 기능은 앞차와 일정 거리를 유지하며 자율 주행을 하는 기능이다. 기존의 SCC 기능은 가속성, 앞차와의 거리, 등을 모두 운전자가 스스로 설정해야 했으며 속도 등이 세밀하지 않아서 운전자는 불안감을 느껴서 이 기능을 꺼리는 경우도 있었다. 그러나 이번에 현대기아차가 개발한 이 기능은 먼저 카메라와 센서 등이 다양한 상황에서 발생되는 정보를 계속 수집하여 그것으로 운전자의 주행 성향을 파악하기에 더 신뢰하고 안정적이게 자율 행을 할 수 있다. 
주행 성향에는 크게 반응성과 가속성 그리고 앞차와의 거리가 있다. 주행 성향에 대한 정보는 꾸준히 센서를 통해 업데이트되기에 운전자가 덜 불안해하며 운전할 수 있도록 도와준다. 

**5. 느낀 점**  
짧게 현재 인공지능이 어느 정도까지 발전해 왔으며 앞으로 더욱 발전할 것이란 것을 알아보았다. 아직은 막연하게 느껴지지만 앞으로 이 분야에 대해 깊게 공부할 것이란 생각에 설레기도 한다. 네 가지 분야 모두 다 매력적으로 느껴지며 모두 흥미롭게 느껴진다. 앞으로 어떤 분야에 대해 더 심도 있게 공부하게 될지는 모르지만 기본적으로 갖추어야 할 인공지능 기반의 언어와 공부들을 더 하고 싶다는 동기가 되었다.
